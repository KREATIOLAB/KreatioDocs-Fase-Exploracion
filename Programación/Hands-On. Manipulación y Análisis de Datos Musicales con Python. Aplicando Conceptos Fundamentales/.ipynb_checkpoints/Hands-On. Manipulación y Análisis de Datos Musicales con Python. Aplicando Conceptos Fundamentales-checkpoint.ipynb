### Step 2: Generación del Conjunto de Datos

En este segundo paso, profundizaremos en la *historia detrás de los datos*. Exploraremos cómo se generó el conjunto de datos musical simulado que estamos a punto de analizar. Descubriremos el contexto detrás de estos datos ficticios y cómo se relacionan con el mundo de la música. Este paso nos ayudará a comprender la naturaleza de los datos y por qué son tan interesantes de analizar.

<div class="alert alert-success" style="color:black">

**Tips:** 

En el mundo del análisis de datos, crear conjuntos de **datos sintéticos** es una práctica esencial. Nos permite desarrollar y probar algoritmos, modelos y técnicas de análisis de datos en un entorno controlado antes de aplicarlos a datos del mundo real. En este paso, descubriremos cómo se generó este conjunto de datos ficticios y cómo se relaciona con la práctica habitual en el análisis y ciencia de datos. Comprenderemos el contexto detrás de estos datos sintéticos y por qué son una herramienta valiosa en nuestro arsenal de análisis de datos.

</div>

```python
# Crear un conjunto de datos vacío
data = []

# Definir el número de filas que deseas (por ejemplo, 200)
num_filas = 200

# Definir listas de valores para cada columna
titulos = ["Canción " + str(i) for i in range(1, num_filas + 1)]
artistas = ["Artista " + str(i) for i in range(1, num_filas + 1)]
generos = ["Pop", "Rock", "Hip-Hop", "Electrónica", "Jazz"] * (num_filas // 5)

# Duración aleatoria entre 120 y 420 segundos
duracion = [random.randint(120, 420) for _ in range(num_filas)]

# Popularidad aleatoria entre 50 y 100
popularidad = [random.randint(50, 100) for _ in range(num_filas)]

# Año de lanzamiento aleatorio entre 1960 y 2022
lanzamiento = [random.randint(1960, 2022) for _ in range(num_filas)]

# Instrumentos aleatorios de una lista predefinida
instrumentos = ["Guitarra", "Batería", "Sintetizador", "Bajo", "Voz", "Piano", "Saxofón", "Percusión"] * (num_filas // 8)

# Construir el conjunto de datos fila por fila
for i in range(num_filas):
    fila = [titulos[i], artistas[i], generos[i], duracion[i], popularidad[i], lanzamiento[i], instrumentos[i]]
    data.append(fila)

# Crear un DataFrame de pandas
df = pd.DataFrame(data, columns=["Título de la Canción", "Artista", "Género", "Duración (segundos)", "Popularidad", "Año de Lanzamiento", "Instrumentos Utilizados"])

# Guardar el DataFrame como un archivo CSV
df.to_csv("conjunto_de_datos_musica_generado.csv", index=False)
```

```python
# Filtrar canciones de un género musical específico
genero_seleccionado = 'Pop'
canciones_pop = df[df['Género'] == genero_seleccionado]
canciones_pop
```

```python
# Filtrar canciones de un año específico
año_seleccionado = 2000
canciones_2000 = df[df['Año de Lanzamiento'] == año_seleccionado]
canciones_2000
```

6. Manipulación de columnas: Explicaremos cómo agregar, eliminar o modificar columnas en el DataFrame según las necesidades del análisis:

```python
# Agregar una nueva columna calculada (por ejemplo, popularidad promedio por género)
df['Popularidad Promedio'] = df.groupby('Género')['Popularidad'].transform('mean')

# Eliminar una columna
df = df.drop(columns=['Instrumentos Utilizados'])
df.head()
```

7. Tratamiento de valores faltantes: Si el conjunto de datos generado contiene valores faltantes intencionadamente, enseñaremos cómo detectarlos y manejarlos, ya sea eliminando filas o imputando valores:

```python
# Imputar valores faltantes en la columna 'Duración (segundos)' con la media
df['Duración (segundos)'].fillna(df['Duración (segundos)'].mean(), inplace=True)

# Eliminar filas con valores faltantes
df.dropna(subset=['Género'], inplace=True)
df.head()
```

8. Ejemplos prácticos: Proporcionaremos ejemplos prácticos de preguntas de análisis de datos que se pueden abordar en esta etapa inicial de exploración:

```python
# ¿Cuántas canciones de cada género hay en el conjunto de datos?
conteo_generos = df['Género'].value_counts()
conteo_generos
```

<div class="alert alert-success" style="color:black">

**Nota:**
    

En el **Step 1**: Limpieza de datos, nos enfocamos en asegurarnos de que nuestro conjunto de datos esté en las mejores condiciones posibles para el análisis. Esto implica identificar y tratar ciertos problemas comunes que pueden surgir en los datos.

**Valores faltantes**: Uno de los problemas más comunes es la presencia de valores faltantes en el conjunto de datos. Estos son valores que están ausentes en algunas filas o columnas. En el código, vimos cómo eliminar filas que contienen valores faltantes utilizando `df.dropna()`. Otra estrategia común es reemplazar estos valores faltantes por algún valor predeterminado, como la media o la mediana de la columna.

**Valores atípicos**: Los valores atípicos son valores inusuales o extremos que pueden distorsionar nuestros análisis. No abordamos esto explícitamente en el código, pero es importante reconocerlos y decidir cómo manejarlos. A veces, es necesario eliminarlos si afectan significativamente los resultados.

**Consistencia de datos**: También es esencial garantizar la consistencia de los datos. Esto implica verificar si los valores se alinean correctamente con el tipo de datos esperado en cada columna. Por ejemplo, asegurarse de que las fechas estén en el formato correcto y que las categorías estén escritas de manera uniforme.

**Detección de duplicados**: Otra tarea importante es identificar y eliminar registros duplicados en el conjunto de datos. La presencia de duplicados puede llevar a resultados sesgados en el análisis.

En resumen, la limpieza de datos es una parte fundamental del proceso de preparación de datos. ### Step 2: Transformación de datos

En este paso, nos adentraremos en la transformación de datos. Aplicaremos funciones y estructuras de control para crear nuevas columnas derivadas o modificar las existentes. Por ejemplo, podemos convertir la duración de las canciones de segundos a minutos o categorizar los años de lanzamiento en décadas. Estas transformaciones nos ayudarán a obtener información más significativa y preparar los datos para el análisis estadístico y la visualización.

```python
# Ejemplo de código para convertir la duración de segundos a minutos
df["Duración (minutos)"] = df["Duración (segundos)"] / 60

# Ejemplo de código para categorizar los años de lanzamiento en décadas
def categorizar_decada(anio):
    if 1960 <= anio < 1970:
        return "1960s"
    elif 1970 <= anio < 1980:
        return "1970s"
    # ... continuar con más décadas

df["Década de Lanzamiento"] = df["Año de Lanzamiento"].apply(categorizar_decada)
df.head()
```

<a id="step1S3"></a> 
### Step 1: Aplicación de análisis estadístico

En este primer paso, utilizaremos Python para realizar análisis estadísticos básicos en nuestros datos musicales. Calcularemos estadísticas descriptivas, como la media, la mediana y la desviación estándar, para comprender mejor las características de las canciones en nuestro conjunto de datos. Estos análisis nos proporcionarán información crucial sobre la distribución de las duraciones, la popularidad y otros aspectos de las canciones.

1. Calculamos la duración promedio de las canciones en segundos:

```python
duracion_promedio = df['Duración (segundos)'].mean()
print("Duración promedio de las canciones:", duracion_promedio, "segundos")
```

2. Calculando la popularidad máxima y mínima:

```python
# Encontramos la canción más y menos popular
cancion_mas_popular = df[df['Popularidad'] == df['Popularidad'].max()]
print("Canción más popular:")
cancion_mas_popular[['Título de la Canción', 'Artista', 'Popularidad']]
```

```python
cancion_menos_popular = df[df['Popularidad'] == df['Popularidad'].min()]
print("\nCanción menos popular:")
cancion_menos_popular[['Título de la Canción', 'Artista', 'Popularidad']]
```

3. Calculando la mediana de la popularidad por género musical:

```python
# Calculamos la mediana de la popularidad para cada género musical
mediana_popularidad_por_genero = df.groupby('Género')['Popularidad'].median()
print("Mediana de popularidad por género musical:")
mediana_popularidad_por_genero
```

### Step 2: Introducción a la Visualización de datos

En el segundo paso, nos adentraremos en el mundo de la visualización de datos en Python. Aprenderemos cómo crear gráficos y visualizaciones efectivas utilizando bibliotecas como Matplotlib y Seaborn. Exploraremos diferentes tipos de gráficos, como histogramas, diagramas de dispersión y diagramas de barras, para representar y comunicar los resultados de nuestros análisis de una manera clara y visualmente atractiva.

1: **Histograma de Duración de Canciones**

Una de las formas más simples y efectivas de visualizar datos numéricos es a través de un histograma. Aquí, crearemos un histograma para la columna de "Duración (segundos)" de nuestro conjunto de datos musical. Esto nos permitirá comprender mejor la distribución de las duraciones de las canciones:

```python
# Crear un histograma de la duración de las canciones
plt.hist(df['Duración (segundos)'], bins=20, color='skyblue', edgecolor='black')
plt.title('Histograma de Duración de Canciones')
plt.xlabel('Duración (segundos)')
plt.ylabel('Número de Canciones')
plt.show()
```

<div class="alert alert-info" style="color:black">

**Tips**

Este histograma nos mostrará cómo se distribuyen las duraciones de las canciones en nuestro conjunto de datos y nos dará una idea de si predominan canciones cortas o largas.

</div>

2. **Gráfico de Barras para Géneros Musicales**

Otra forma de visualizar datos es a través de un gráfico de barras. Aquí, crearemos un gráfico de barras para mostrar la cantidad de canciones en cada género musical en nuestro conjunto de datos.

```python
# Contar la cantidad de canciones por género
conteo_generos = df['Género'].value_counts()

# Crear un gráfico de barras
plt.figure(figsize=(10, 6))
conteo_generos.plot(kind='bar', color='salmon')
plt.title('Cantidad de Canciones por Género Musical')
plt.xlabel('Género Musical')
plt.ylabel('Número de Canciones')
plt.xticks(rotation=45)
plt.show()
```

3. **Gráfico de Dispersión entre Duración y Popularidad**:

Un gráfico de dispersión nos permite visualizar la relación entre dos variables numéricas. Aquí, crearemos un gráfico de dispersión para explorar si existe alguna relación entre la duración de las canciones y su popularidad.

```python
# Crear un gráfico de dispersión entre duración y popularidad
plt.figure(figsize=(8, 6))
plt.scatter(df['Duración (segundos)'], df['Popularidad'], alpha=0.5, color='green')
plt.title('Gráfico de Dispersión entre Duración y Popularidad de Canciones')
plt.xlabel('Duración (segundos)')
plt.ylabel('Popularidad')
plt.show()
```

<div class="alert alert-info" style="color:black">

La visualización de datos es una herramienta esencial en el análisis de datos y nos ayuda a comunicar información de manera efectiva.
