## Section 1: Tipos de Datos

Aquí se describen algunos de los tipos de datos numéricos más importantes de Numpy que encuentras con frecuencia. Numpy admite numerosos tipos de datos, tal vez incluso más que Python mismo.

<a id="s11"></a> 
### Step 1: Tipos de Datos Básicos

Los tipos de datos más básicos son los siguientes: **entero** (ejemplo: 1, 2, -10), **flotante** (ejemplo: 1.1, 3.24, -7.00111), **complejo** (ejemplo: 1 + 2j, 2.1 + 3j, -2 + 1.4j) y **booleano** (ejemplo: True, False). Puedes usar Numpy para convertir elementos de Python de muchas maneras diferentes, como cambiar el tipo de array a otro específico. Comencemos con los siguientes ejemplos:

```python
# Importamos la librería de Numpy con el alias np
import numpy as np

# Definimos un número entero
a = 10
print("Tipo, antes de la conversión: ", type(a))

# Cambiamos el tipo de dato a float con NumPy
b = np.float64(a)
print("Tipo, después de la conversión: ", type(b))

# Cambiamos el tipo de dato a float con Python
c = float(a)
print("Tipo, después de la conversión: ", type(c))
```

<a id="s12"></a> 
### Step 2: Conversión de tipo

Ahora, convirtamos una lista en un array de NumPy. Por ahora, centrémonos solo en los tipos de datos y conversiones. Más adelante en este capítulo, verás cómo definir arrays de NumPy en detalle.

```python
# Definimos una lista de Python
a = [1 , 2 , 1.3]
print("Tipo de 'a': ", type(a))

# Convertimos la lista en un array de NumPy de tipo np.int32 (Integer( -2147483648 to 2147483647))
b = np.int32 (a)
print("\nArray 'b': ", b)
print("Tipo de array 'b': ", type(b))
print("Data type de 'b': ", b.dtype)

# Convertimos la lista en un array de NumPy de tipo np.float32 (mismo que float de Python)
c = np.float32(a)
print("\nArray 'c': ", c )
print("Tipo de array 'c': ", type(c))
print("Data type de 'c'", c.dtype)
```

<a id="s13"></a> 
### Step 3

Utilizamos el método `.dtype` de NumPy para comprender cuál es el tipo de datos dentro del array. La forma recomendada de cambiar el tipo de un array de NumPy es mediante el uso del método `.astype()`. Echa un vistazo al siguiente ejemplo:

```python
# Definimos una lista de Python
a = [1 , 2 , 4]
print("Type 'a': ", type(a))

# Convertimos la lista en un array de NumPy de tipo np.int32 (Integer( -2147483648 to 2147483647))
b = np.int32(a)
print("\nArray 'b': ", b)
print("Tipo de array 'b': ", type(b))

# Convertimos la lista en un array de NumPy de tipo np.float32 (mismo que float de Python)
c = b.astype(np.float32)
print("\nArray 'c': ", c)
print("Tipo de array 'c': ", type(c))
```

## Section 2: Definiendo un array de Numpy

Un array es básicamente una cuadrícula de valores unidimensional o multidimensional. En un array de NumPy, en particular, todos los valores son del mismo tipo (entero, flotante). ¿Cómo vamos a definir un array de NumPy? Para un array de NumPy, tenemos las siguientes definiciones:

* **Rango**: El número de dimensiones que tiene un array.
* **Forma**: Una tupla que indica el número de elementos en cada dimensión. Por ejemplo, la forma de un array siendo (2, 4, 10) indica que tenemos un array tridimensional que tiene 2, 4 y 10 elementos en la primera, segunda y tercera dimensión, respectivamente. 

<a id="s21"></a> 
### Step 1: Crear un array a partir de una lista

Ahora, comencemos con Python. Empezaremos con el enfoque más común. Vamos a definir y crear un array de NumPy a partir de una lista:

```python
# Definimos una lista de Python
mi_lista = [1 , 2 , 4 , 8]

# Creamos un array de Numpy a partir de una lista
numpy_array = np.array(mi_lista)
print("Array:", numpy_array)
```

<a id="s22"></a> 
### Step 2

Qué pasa si no definimos una lista y simplemente ingresamos los números como se muestra a continuación:

```python
# Ingresar los números de manera directa
numpy_array = np.array((1, 2, 4, 8))
print("Array:", numpy_array)
```

<a id="s23"></a> 
### Step 3

Ahora, definamos una matriz bidimensional:

```python
# Ingresar los números de manera directa
row1 = [2, 4, 6, 8]
row2 = [1, 3, 5, 7]
numpy_array = np.array([row1, row2])
print("Array:", numpy_array)

# Obtenemos la dimensión
print("Dimensión:", numpy_array.shape)
```

<div class="alert alert-warning" style="color:black">
Nota: Echemos un vistazo al código anterior una vez más. Hemos definido una matriz. El argumento dentro de np.array es una lista en la que cada uno de sus elementos es otra lista. Las listas internas, denotadas como `row1` y `row2`, forman las filas de la matriz ¡y DEBEN tener el mismo tamaño! El método `.shape` se utiliza para devolver la forma del array NumPy. La salida anterior muestra que tenemos una matriz con dos filas y cuatro columnas.</div>

<a id="s24"></a> 
### Step 4: Funciones especiales

Podemos crear matrices NumPy utilizando algunas funciones especiales de NumPy. He utilizado un par de ellas a continuación para su referencia.

```python
### Definiendo arrays usando funciones especiales ###
# Argumentos:
    # shape: La dimensión del arreglo de numpy
    # dtype: Si queremos especificar el tipo de dato (no obligatorio)

# Definiendo un array de todos ceros
cerosArray = np.zeros(shape = (3, 5), dtype = np.int16)
print("cerosArray:", cerosArray)

# Definiendo un array de todos unos
unosArray = np.ones(shape = (3, 5), dtype = np.float32)
print("\nunosArray:", unosArray)

# Definiendo un array con un número específico
fullArray = np.full(shape = (3, 5), fill_value = 4.2, dtype = np.float64)
print("\nfullArray:", fullArray)
```

<a id="s25"></a> 
### Step 5

Una de las funciones especiales más importantes es `np.arange`, que es similar a la función incorporada range de Python. Un ejemplo de cómo usar `np.arange` es el siguiente:

```python
# Definiendo un array de NumPy usando np.arrage
# np.arrage define un intervalo de números
# Argumentos:
    # start: Comienzo del intervalo.
    # stop: Final del intervalo.
    # step: Amplitud del intervalo.
# NOTA : El intervalo incluye el número 'start' pero NO incluye el número 'stop'
arr = np.arange (start = 3, stop = 10, step = 2)
print("Array:", arr)
```

<a id="s26"></a> 
### Step 6: Funciones universales

NumPy, para realizar operaciones elemento a elemento, proporciona algunas funciones universales. Echa un vistazo a algunas de las funciones universales demostradas a continuación:

```python
# Definimos dos arrays de Numpy
v = np.array([1, 3, 5])
w = np.array([2, 4, 6])

# Operación exponencial
print("e^v=", np.exp(v))

# La raíz cuadrada de un array
print("v^{1/2}=", np.sqrt(v))

# Sumando dos vectores
print("v+w=", np.add(v, w))

# Seno y Coseno de un array
print("Sin(v)=", np.sin(v))
print("Cos(v)=", np.cos(v))
```

<a id="s27"></a> 
### Step 7: Arrays con random

A veces, se desea definir una matriz con números aleatorios. Hay muchas formas de hacer esta tarea dependiendo del tipo de números aleatorios que queramos utilizar. Por ejemplo, ¿queremos generar enteros aleatorios, flotantes, etc.? Echa un vistazo a los enfoques que se muestran a continuación:

```python
### Definiendo arrays con elementos aleatorios ###
# Argumentos :
    # size : El tamaño del array

# Definimos un array aleatorio
randArray = np.random.random(size = (2, 3))
print("randArray:", randArray)

# Definimos un array aleatorio con números enteros
    # high : El elemento más alto que se permite generar
    # low: El elemento más bajo que se permite generar
randintArray = np.random.randint(low = 1, high = 5, size = (2 ,3))
print("randintArray:", randintArray)
```

## Section 3: Operaciones Matemáticas Básicas

Primero cubramos las operaciones aritméticas básicas con algunos ejemplos:

<a id="s31"></a> 
### Step 1: Adición

```python
# Creamos dos arrays
a = np.array([[1 ,5], [0 ,8]], dtype = np.float32)
b = np.array([[2 ,1], [5 ,2]], dtype = np.float32)
print("a=", a)
print("b=", b)

# Adición elemento por elemento
# a + b y np.add(a, b) es lo mismo.
# Puedes asegurarte si a + b y np.add(a, b) es lo mismo usando ‘assert‘.
# El método .all() comprueba si todos los elementos de la matriz son True.
print("a + b= ", a + b)
assert((a + b == np.add (a, b)).all())
```

<a id="s32"></a> 
### Step 2: Resta

```python
# Resta elemento por elemento
print("a - b=", a - b)
assert((a - b == np.subtract(a, b)).all())
```

<a id="s33"></a> 
### Step 3: Multiplicación

```python
# Multiplicación elemento por elemento
print("a * b=", a * b)
assert((a * b == np.multiply(a, b)).all())
```

<a id="s34"></a> 
### Step 4: División

```python
# División elemento por elemento
print("a / b=", a / b)
assert((a / b == np.divide(a, b)).all())
```

<a id="s35"></a> 
### Step 5: Cuadrado

```python
# El cuadrado de elemento por elemento
print("a ^ 2=", np.square(a))
assert((a ** 2 == np.square(a)).all())
```

## Section 4: Manipulación de Arrays

Después de familiarizarse con los arrays de NumPy, ahora es el momento de aprender cómo manipular los arrays y conocer algunas técnicas.

<a id="s41"></a> 
### Step 1: Indexing

Primero, vamos a definir e ingresar en un array:

```python
# Creamos una matriz
A = np.array([[2, 1, 3, 4], [5, 2, 9, 4], [5, 2, 10, 1], [2, 2, 11, -1]])
print("A:\n", A)

# Extraemos las primeras dos filas
# Recuerda que 0:2 en indexing significa {0 ,1} y NO incluye el 2!
print("\nLas primeras dos filas: \n", A[0:2, :])

# Extraemos las primeras tres filas y las últimas dos columnas.
# -2: significa el segundo desde el último hasta el final
# 0:3 significa {0 ,1 ,2}
print("\nLas primeras tres filas y las últimas dos columnas: \n", A[0:3 , -2:])

# Vamos a señalar un elemento
# La segunda fila (índice 1) y la tercera columna (índice 2)
# Recuerda que el indexado en Python comienza desde cero
print("\nLa segunda fila y la tercera columna: \n", A[1 ,2])
```

<a id="s42"></a> 
### Step 2

Veamos el código anterior. Puedes utilizar indexación de segmentación seleccionando un rango de índices o sólo utilizar índices enteros. Podemos simplemente combinar ambos, pero podría haber una diferencia en la clasificación de la matriz de salida. Revisa el ejemplo a continuación:

```python
# Creamos una matriz
A = np.array([[2, 1, 3, 4], [5, 2, 9, 4], [5, 2, 10, 1], [2, 2, 11, -1]])
print("A:\n", A)

# Extraemos la primera fila y todas las columnas usando dos enfoques.
print("\nLa primera fila con slice indexing: \n", A[0:1 ,:]) # Slice indexing
print("\nLa primera fila con integer indexing: \n", A[0 ,:]) # Integer indexing
print("\nLa dimensión de la primera fila con slice indexing: \n", A[0:1 ,:].shape) # Slice indexing
print("\nLa dimensión de la primera fila con integer indexing: \n", A[0 ,:].shape) # Integer indexing
```

Si observas los resultados, la forma de la matriz sería diferente. Básicamente, con la indexación de segmentación, tenemos una matriz de rango 2 y con la indexación entera, tendremos una matriz de rango 1. Ten cuidado con esta diferencia cuando estés trabajando con la indexación de NumPy. Más adelante hablaremos en detalle sobre el concepto de rango de la matriz.

<a id="s43"></a> 
### Step 3: Shaping

La pregunta es ¿cómo funcionan las operaciones de *shaping*? Arriba teníamos la matriz A de tamaño (3,4) con un total de 12 elementos (3×4). Cuando usamos `np.reshape`, el orden predeterminado de NumPy es "estilo C", es decir, el índice más a la derecha "cambia más rápido" para la operación de procesamiento. Usemos el ejemplo anterior de usar `.ravel()` para aplanar la matriz: El primer elemento es obviamente A[0,0] y el siguiente es A[0,1]. El proceso y la creación de la nueva matriz es como se muestra a continuación al usar `.ravel()`:

```python
# Creamos una matriz
A = np.array([[2, 1, 3, 4], [5, 2, 9, 4], [5, 2, 10, 1]])
print("A:\n", A)
print("\nDimensión de A:\n", A.shape)

# Redimensiona A a la nueva forma de (2, 6)
B = A.reshape(2, 6)
print("\nB:\n", B)

# Redimensiona A a la nueva forma de (2, x)
# Si usamos -1, la dimensión restante será elegida automáticamente.
C = A.reshape(4, -1)
print("\nC: \n", C)

# Operación Flatten
print("\nFlatten A:\n", A.ravel())
```
